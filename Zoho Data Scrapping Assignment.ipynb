```python
import requests

ACCESS_TOKEN = access_token
CLIQ_BASE = "https://cliq.zoho.com/api/v2"  # change domain if you use .eu, .in, etc.

def get_all_users():
    all_users = []
    next_token = None

    while True:
        params = {
            "limit": 100,          # max allowed
            "fields": "all",       # optional: get all fields
        }
        if next_token:
            params["next_token"] = next_token

        headers = {
            "Authorization": f"Zoho-oauthtoken {ACCESS_TOKEN}"
        }

        resp = requests.get(f"{CLIQ_BASE}/users", headers=headers, params=params)
        resp.raise_for_status()
        data = resp.json()

        # users are in "data" array
        batch = data.get("data", [])
        all_users.extend(batch)

        # pagination flags
        has_more = data.get("has_more", False)
        next_token = data.get("next_token")

        if not has_more or not next_token:
            break

    return all_users

if __name__ == "__main__":
    users = get_all_users()
    print("Total users fetched:", len(users))
    # Example: print first few user emails
    for u in users[:10]:
        print(u.get("email_id"), "-", u.get("display_name"))
```

```python
import requests
import pandas as pd

# Replace this with your valid access token
ACCESS_TOKEN = access_token
CLIQ_BASE = "https://cliq.zoho.com/api/v2"  # use .eu or .in if your account is there

all_users = []
next_token = None

while True:
    params = {"limit": 100}  # 100 = max per request
    if next_token:
        params["next_token"] = next_token

    headers = {"Authorization": f"Zoho-oauthtoken {ACCESS_TOKEN}"}

    # Call the Zoho Cliq API
    response = requests.get(f"{CLIQ_BASE}/users", headers=headers, params=params)
    response.raise_for_status()
    data = response.json()

    # Add this batch of users
    all_users.extend(data.get("data", []))

    # Stop when there’s no more data
    if not data.get("has_more"):
        break

    # Otherwise, continue with next_token
    next_token = data.get("next_token")

# Convert to DataFrame
df = pd.DataFrame(all_users)

# Optional: show a preview
print(df.head())

# Save all users to CSV
df.to_csv("user3.csv", index=False)
print(f"✅ Done! Saved {len(df)} users to user3.csv")
```
